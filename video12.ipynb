# Sadece belli columnları almak istersek

```python
import pandas as pd
```

```python
drinks = pd.read_csv("data/drinks.csv")
```

```python
drinks.columns
```

```python
##continent columunu siliyoruz, ama asıl datayı değiştirmeden: on fly
drinks.drop('continent', axis=1).head()
```

```python
##2. rowu siliyoruz, axis 0 oldu, çünkü rowlarla ilgileniyoruz. 
drinks.drop(2, axis=0).head()
```

```python
# mean değerini kullanıyoruz.
drinks.mean()
# tüm sayısal değerlerin ortalamasını bulacaktır.
drinks.mean(axis=0) # aynı sonucu verecektir. ya da 
drinks.mean(axis='index') # aynı sonucu verecektir.
```

```python
# her row daki sayısal değ'columns'rlerin ortalaması için (bu veri seti için anlamlı olmasa da yapabildiğini göstermek için)
drinks.mean(axis=1)
# ya da 
drinks.mean(axis='columns')
```
