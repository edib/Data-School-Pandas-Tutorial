# When should I use the "inplace" parameter in pandas?
https://www.youtube.com/watch?v=XaCSdr7pPmY

```python
import pandas as pd
```

```python
ufo = pd.read_csv("data/ufo.csv")
```

```python
#datamız neye benziyor bakalım
ufo.head()
```

```python
# city columnunu kaldırıyoruuz.
ufo.drop('City',axis=1).head()
```

```python
# ama yerinde hala duruyor
ufo.head()
```

```python
# kalıcı olarak kaldırmak için.
ufo.drop('City',axis=1, inplace=True)
ufo.head()
```

```python
# mesela boş olanları kaldırmak isteyebiliriz ve ne kadarını kaldırdığımıza böyle bakabiliriz.
ufo.dropna(how="any").shape
```

```python
# aslında hala yerindedir. 
ufo.shape
```

```python
# eğer orjinal datada değiştirmek istersek
ufo.dropna(how="any", inplace=True)
ufo.shape
```

```python
# inplace kullanabileceğimiz birçok df methodu vardır. Bunların default değerleri her zaman inplace=False tur.
ufo.sort_values('Time',inplace=True)
# yada aşağıdakini kullanabiliriz. 2. nin kötü tarafı bellekte 2.si kadar yer işgal etmesidir. 
ufo = ufo.sort_values('Time') 
ufo.set_index('Time',inplace=True)
```
